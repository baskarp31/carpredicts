```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv('Car-Price-Prediction.csv')
df
```

```python
inputs=df.drop(['Car_Name','Owner','Seller_Type'],axis='columns')
inputs
```

```python
#Target variable choosse...
target=df.Selling_Price
target
```

```python
#encoing label column  Fuel_Type 	 Transmission	
from sklearn.preprocessing import LabelEncoder
Numerics=LabelEncoder()
```

```python
#new encode column created..

inputs['Fuel_Type_n']=Numerics.fit_transform(inputs['Fuel_Type'])
inputs['Transmission_n']=Numerics.fit_transform(inputs['Transmission'])
inputs
```

```python
#drop label column values. Fuel_Type 	 Transmission
inputs_n=inputs.drop(['Fuel_Type','Transmission','Selling_Price'],axis='columns')
inputs_n
```

```python
# Starts : linear regression 
model=linear_model.LinearRegression()
```

```python
#Model will Training ...
model.fit(inputs_n,target)
```

```python
#prediction car price
pred=model.predict([[2013,9.54,43000,1,1],[2016,5.90,43000,2,2]])
print(pred)
```

```python
### Create a Pickle file using serialization 
import pickle
pickle_out = open("carprice.pkl","wb")
pickle.dump(model, pickle_out)
pickle_out.close()
```
